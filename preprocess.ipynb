```python
import pandas as pd
```

```python
df = pd.read_csv('spotify_data.csv')
```

```python
del df['Unnamed: 0']
```

```python
df.head()
```

```python
drop_genres = ['black-metal', 'death-metal', 'gospel', 'goth', 'heavy-metal', 
               'indie-pop', 'k-pop', 'metal', 'metalcore', 'opera', 'french',
              'german', 'indian', 'salsa', 'samba', 'spanish', 'swedish']
```

```python
df = df[~df['genre'].isin(drop_genres)]
```

```python
df['genre'].unique()
```

```python
songs = pd.read_csv('audio_features.csv')
```

```python
songs.head()
```

```python
format = ['id', 'danceability', 'energy', 'loudness', 'speechiness', 'acousticness', 'instrumentalness', 'liveness', 'valence', 'tempo']
```

```python
songs = songs[format]
```

```python
songs.head()
```

```python
df_format = ['track_id', 'danceability', 'energy', 'loudness', 'speechiness', 'acousticness', 'instrumentalness', 'liveness', 'valence', 'tempo']
```

```python
df_format
```

```python
df = df[df_format]
```

```python
df.head()
```

```python
df = df.rename(columns={'track_id': 'id'})
```

```python
df.head()
```

```python
sampled_df = df.sample(n=10000, replace=False, random_state=42)
```

```python
sampled_df
```
